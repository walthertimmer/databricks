```python
import yfinance as yf


def get_quote_info(ticker):
    print(f"collecting data for: {ticker}")
    ticker_info = yf.Ticker(f"{ticker}")
    if debug:
        print(ticker_info.info)

    

    #ask_price = ticker_info.info['ask']
    # dividend_rate = ticker_info.info['dividendRate']
    # dividend_yield = ticker_info.info['dividendYield']
    # payout_ratio = ticker_info.info['payoutRatio']
    # pe_rate = ticker_info.info['forwardPE'] # price / earnings rate
    # fiveYearAvgDividendYield
    # beta
    # trailingPE
    # marketCap
    # currency
    # profitMargins
    # bookValue
    # priceToBook
    # earningsQuarterlyGrowth
    # trailingEps
    # forwardEps
    # pegRatio
    # enterpriseToRevenue
    # enterpriseToEbitda
    # shortName
    # currentPrice
    # recommendationMean
    # recommendationKey
    # quickRatio
    # currentRatio
    # debutToEquity
    # revenuePerShare
    # returnOnAssets
    # returnOnEquity
    # earningsGrowth
    # revenueGrowth
    # grossMargins
    # ebitdaMargings
    # operatingMargings
    # trailingPegRatio
    return ticker_info.info
```

```python
# stocks to analyze
stocks = [
    ('AAPL') #Apple
    ,('MSFT') # Microsoft
    ,('NVDA') # Nvidia
    ,('AMD') # Advanced Micro Devices Inc.
    ,('BUD') # Anheuser-Busch InBev
    ,('V') # VISA
    ,('AMZN') # Amazon
    ,('META') # Facebook
    ,('WMT') # Walmart
    ,('NSRGY') # Nestle
    ,('UNA.AS') # Unilever
    ,('DANOY') # Danone
    ,('OR.PA') # L'oreal
    ,
]
```

```python
# maybe build in a wait to not irritate yahoo?
import pandas as pd

debug = False
results = []

for ticker in stocks:
    result = get_quote_info(ticker)
    results.append(result)

# only create dataframe at the end to speedup process
df = pd.DataFrame(results)
```
