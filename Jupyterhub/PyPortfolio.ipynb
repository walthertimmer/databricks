```python
import yfinance as yf

# get current info
def get_quote_info(ticker):
    print(f"collecting data for: {ticker}")
    ticker_info = yf.Ticker(f"{ticker}")
    if debug:
        print(ticker_info.info)

    return ticker_info.info
```

```python
import yfinance as yf
import pandas as pd

# get history
def get_quote_history(ticker):
    print(f"collecting history data for: {ticker}")
    ticker = yf.Ticker(f"{ticker}")
    ticker_history = ticker.history(period="max",interval="1mo",actions=False)

    ticker_history = ticker_history[['Close']]
    if debug:
        print(ticker_history.columns)
        print(ticker_history)

    # get prices for minus 1, 3, 5, 10, 15, 20 year
    df_history = pd.DataFrame(ticker_history)
    display(df_history)

get_quote_history("AAPL")
```
